```python
from h2_therm import H2_Combustion
from h2_flame import *
from h2_comps import Gas, Orifice
from h2_therm import AbelNoble
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
H2 = Gas(AbelNoble(), T = 287.8, P = 13420000)
air = Gas(AbelNoble(b = 0, MW = 28.), T = 288.15, P = 101325.)
o = Orifice(.00356)
```

```python
h2chem = H2_Combustion(Treac = air.T)
```

```python
fl = Flame(H2, o, air, h2chem = h2chem, y0 = 0.5, lamf = 1.24, nnmodel = 'Birch2')
```

```python
fl.solve();
```

```python
fl.iso_heat_flux_plot()
```
